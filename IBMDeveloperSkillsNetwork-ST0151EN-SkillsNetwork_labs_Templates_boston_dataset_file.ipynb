Q6: Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)

```python
import scipy.stats
from scipy.stats import ttest_ind
```

Charles River dummy variable (1 if tract bounds river; 0 otherwise)
Use Î± = 0.05

```python
scipy.stats.ttest_ind(boston_df['MEDV'], boston_df['CHAS'])
```

Conclusion: p-value is much smaller than the significance level of 0.05, we can reject the null hypothesis and conclude that there is a significant difference in the median values of houses bounded by the Charles river

Q7: Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)

```python
from scipy.stats import levene
scipy.stats.levene(boston_df[boston_df['age_group'] == "35 years and younger"]['MEDV'],
                   boston_df[boston_df['age_group'] == "between 35 and 70 years"]['MEDV'], 
                   boston_df[boston_df['age_group'] == "70 years and older"]['MEDV'], 
                   center='mean')
```

If the p-value is less than the significance level (usually 0.05), we can reject the null hypothesis and conclude that there is a significant difference in the median values of houses for at least one of the age groups. If the p-value is greater than the significance level, we fail to reject the null hypothesis and conclude that there is no significant difference in the median values of houses for the different age groups.

```python
from scipy.stats import f_oneway
```

```python
thirty_five_and_young= boston_df[boston_df['age_group'] == "35 years and younger"]['MEDV']
btw_thirty_five_and_seventy = boston_df[boston_df['age_group'] == "between 35 and 70 years"]['MEDV']
seventy_and_older = boston_df[boston_df['age_group'] == "70 years and older"]['MEDV']
```

```python
scipy.stats.f_oneway(thirty_five_and_young,btw_thirty_five_and_seventy,seventy_and_older)
```

Conclusion: There is a significant difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)

Q:10 What is the impact of an additional weighted distance  to the five Boston employment centres on the median value of owner occupied homes? (Regression analysis)

```python
# X is the input variables (or independent variables)
X = boston_df['DIS']

# y is the target/dependent variable
y = boston_df['MEDV']

# add an intercept (beta_0) to our model
X = sm.add_constant(X) 

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

# Print out the statistics
model.summary()
```

 p < 0.05 there is evidence of correlation between DIS (weighted distances to five Boston employment centres) and MDEV ( Median value of owner-occupied homes)
